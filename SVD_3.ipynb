```python
import pandas as pd
import numpy as np
from scipy.linalg import svd
```

```python
df = pd.read_csv("iris.data")
df = df.replace(to_replace=["Iris-setosa","Iris-versicolor","Iris-virginica"],value=[0,1,2])
features = df[df.columns[:-1]] 
ndarray = features.to_numpy()
print(ndarray)
```

```python
U, sigma, VT = svd(features)
print("U: \n", U)
print("sigma: \n", sigma)
print("V transpose: \n", VT)
```

```python
# sigma should be a diagonal matrix therefore we should populate with 0
sigma_matrix = np.zeros((ndarray.shape[0], ndarray.shape[1]))
print(sigma_matrix)
```

```python
sigma_matrix[:ndarray.shape[1], :ndarray.shape[1]] = np.diag(sigma)
print(sigma_matrix)
```

```python
B = U.dot(sigma_matrix.dot(VT))
print(B)
```
