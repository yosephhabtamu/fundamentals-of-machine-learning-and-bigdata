```python
import numpy as np
import pandas as pd
from numpy.linalg import eigh,norm
```

```python
df = pd.read_csv("iris.data")
df = df.replace(to_replace=["Iris-setosa","Iris-versicolor","Iris-virginica"],value=[0,1,2])
features = df[df.columns[:-1]] 
```

```python
ndarray = features.to_numpy()
ndarray
```

```python
ndarray.T@ndarray
```

```python
eigen_value, eigen_vector = eigh(ndarray.T@ndarray)
```

```python
eigen_vector
```

```python
u0 = ndarray@eigen_vector[:,0]/norm(ndarray@eigen_vector[:,0])
u1 = ndarray@eigen_vector[:,1]/norm(ndarray@eigen_vector[:,1])
u2 = ndarray@eigen_vector[:,2]/norm(ndarray@eigen_vector[:,2])
u3 = ndarray@eigen_vector[:,3]/norm(ndarray@eigen_vector[:,3])
```

```python
U = np.array([u0,u1,u2,u3]).T
print(U)
```

```python
sigma = np.round(U.T@ndarray@eigen_vector, decimals=5)
print(sigma)
```
