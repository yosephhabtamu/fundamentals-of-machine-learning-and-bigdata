```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("iris.data")
df = df.replace(to_replace=["Iris-setosa","Iris-versicolor","Iris-virginica"],value=[0,1,2])
print(df)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df[df.columns[:-1]])
```

```python
scaled_data = scaler.transform(df[df.columns[:-1]])
```

```python
from sklearn.decomposition import TruncatedSVD
```

```python
svd = TruncatedSVD(n_components=2)
```

```python
svd.fit(scaled_data)
```

```python
x_svd = svd.transform(scaled_data)
print(x_svd)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_svd[:,0], x_svd[:,1], c = df['class'])
plt.xlabel("first pricincipal component")
plt.ylabel("second pricincipal component")
plt.plot()
```
